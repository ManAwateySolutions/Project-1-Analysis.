### Filling in the null values using the method:  df.fillna()
#### I am going to fill in the null values with with the respective mean values of the various columns. This is because, I deem it fit that, the best set of values to replace any null value in the various columns will be to use the average values of those columns.

```python
data["fixed_acidity"] = data["fixed_acidity"].fillna(mean_FA)
data["volatile_acidity"] = data["volatile_acidity"].fillna(mean_VA)
data["citricAcid"] = data["citricAcid"].fillna(mean_CA)
data["residual_sugar"] = data["residual_sugar"].fillna(mean_RS)
data["chlorides"] = data["chlorides"].fillna(mean_CL)
data["free_sulfur_dioxide"] = data["free_sulfur_dioxide"].fillna(mean_FS)
data["total_sulfur_dioxide"] = data["total_sulfur_dioxide"].fillna(mean_TS)
data["density"] = data["density"].fillna(mean_DS)
data["sulphates"] = data["sulphates"].fillna(mean_SP)
data["alcohol"] = data["alcohol"].fillna(mean_AL)
data["quality"] = data["quality"].fillna(mean_QL)
data["pH"] = data["pH"].fillna(mean_PH)
```
